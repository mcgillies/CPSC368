```python
import os 
import pandas as pd
import numpy as np
```

##### Functions were not used here as each dataset has variable conditions that need to be checked/applied.  For example, for the temperature dataset we need to convert the date into year and month columns for merging  purposes, and for generating the insert statements we need different columns, along with different syntax depending on if the column is numerical vs categorical. Therefore we thought that creating functions would be just as if not more strenuous then performing the operations manually.

```python
def clean_df(df):
    # Input is pd.DataFrame
    # Fills na value with "NULL" and filters years to be 2012-2013
    # Return pd.dataframe
    df = df.fillna("NULL")
    df = df[(df['Year'] > 2005) & (df['Year'] < 2014)]
    return df
```

```python
def print_specification_statement(df_name):
    ## Input is String
    ## Prints the statement specifying which dataframe we are inserting to
    ## Returns none
    df_name = str(df_name)
    sql_file.write("\n")
    sql_file.write("--now add in " + df_name + " data")
    sql_file.write("\n")
    
```

```python
## Reading in temperature dataset and filtering/creating columns as necessary
et = pd.read_csv("data/earth_temp_updated.csv")
et['Year'] = et['dt'].str.slice(0,4)
et['Year'] = et['Year'].astype(int)
et['Month'] = et['dt'].str[6].astype(int)
et = clean_df(et)
et
```

```python
# Creating sql file to write insert statements to
sql_file = open('cpsc368_ddl15.sql', 'w')
```

```python
print_specification_statement("temperature")
```

```python
# Creating insert statements by iterating through rows
for index, row in et.iterrows():
    dt = str(row["dt"])
    avgtemp = str(row['AverageTemperature'])
    avgtu = str(row['AverageTemperatureUncertainty'])
    line = 'insert into temperature values (' + "'" + dt + "'" + ', ' + avgtemp + ', ' + avgtu + ');'
    sql_file.write(line +"\n")


```

```python
## Reading in power generated dataset and filtering/creating columns as necessary
pc = pd.read_csv("data/BC_Power_Generation_updated.csv")
pc = pc[['source', 'year', 'value']]
pc = pc.rename(columns = {'year':'Year'})
pc['source'] = pc['source'].replace({'BIOMASS / GEOTHERMAL': 'BIOMASS'})
sources = ['BIOMASS', 'OIL', 'NATURAL GAS']
pc = pc[pc['source'].isin(sources)]
pc = clean_df(pc)
pc
```

```python
# Specifying which table we are inserting to
print_specification_statement("power_generated")
```

```python
# Creating insert statements by iterating through rows
for index, row in pc.iterrows():
    source = str(row["source"])
    year = str(row['Year'])
    value = str(row['value'])
    line = 'insert into power_generated values (' + "'" + source + "'" ', ' + year + ', ' + value + ');'
    sql_file.write(line +"\n")
```

```python
## Reading in power consumed dataset and filtering/creating columns as necessary
pcons = pd.read_csv("data/BC_Power_Consumed_updated (1).csv")
pcons = pcons.rename(columns = {'year':'Year'})
pcons = pcons.fillna("NULL")
pcons['source'] = pcons['source'].replace({"OIL PRODUCTS":"OIL", "BIOFUELS & EMERGING ENERGY":"BIOMASS"})
sources = ['BIOMASS', 'OIL', 'NATURAL GAS']
pcons = pcons[pcons['source'].isin(sources)]
pcons = clean_df(pcons)
pcons
```

```python
# Specifying which table we are inserting to
print_specification_statement("power_consumed")
```

```python
# Creating insert statements by iterating through rows
for index, row in pcons.iterrows():
    source = str(row["source"])
    year = str(row['Year'])
    value = str(row['value'])
    line = 'insert into power_consumed values (' + "'" + source + "'" + ', ' + year + ', ' + value + ');'
    sql_file.write(line +"\n")
```

```python
## Reading in precipitation dataset and filtering/creating columns as necessary
precip = pd.read_csv("data/Station_data_cleaned_updated.csv")
precip['MAX_TEMP'] = precip['MAX_TEMP'].replace(' None', np.nan)
precip['MAX_TEMP'] = precip['MAX_TEMP'].astype(float)
precip['ONE_DAY_PRECIPITATION'] = precip['ONE_DAY_PRECIPITATION'].replace(' None', np.nan)
precip['ONE_DAY_PRECIPITATION'] = precip['ONE_DAY_PRECIPITATION'].astype(float)
precip = clean_df(precip)
stations = ['MIS', 'DCN', 'CQM', 'GRP', 'NTY', 'SCA', 'CRU', 'BCK', 'SLK']
precip = precip[precip['Station'].isin(stations)]
precip['Station'].unique()
```

```python
# Specifying which table we are inserting to
print_specification_statement("precipitation")
```

```python
# Creating insert statements by iterating through rows
for index, row in precip.iterrows():
    station = str(row["Station"])
    year = str(row['Year'])
    month = str(row['Month'])
    mintemp = str(row['MIN_TEMP'])
    maxtemp = str(row['MAX_TEMP'])
    pre = str(row['ONE_DAY_PRECIPITATION'])
    line = 'insert into precipitation values (' + "'" + station + "'" + ', ' + year + ', ' + month + ', ' + maxtemp + ', ' + mintemp + ', ' + pre +');'
    sql_file.write(line +"\n")
```

```python
## Reading in population dataset, merging as necessary and filtering/creating columns as necessary
pop_data = pd.read_csv("data/BC_Population_2005-2002_updated.csv")

pop_pg = pop_data.merge(pc, on = 'Year')
pop_pgc = pop_pg.merge(pcons, on = 'Year')
pop_pgc = pop_pgc.drop_duplicates(['Year', 'Region'])
pop_pgc = clean_df(pop_pgc)
pop_pgc
```

```python
# Specifying which table we are inserting to
print_specification_statement("population")
```

```python
# Creating insert statements by iterating through rows
for index, row in pop_pgc.iterrows():
    sourcepg = str(row["source_x"])
    sourcepc = str(row['source_y'])
    year = str(row['Year'])
    region = str(row['Region'])
    munic = str(row['Municipality']).replace("'", "")
    total = str(row['Total'])
    line = f"insert into population values ({year}, {region}, '{munic}', '{total}', '{sourcepg}', {year}, '{sourcepc}', {year});"
    sql_file.write(line +"\n")
```
