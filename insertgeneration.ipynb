```python
import os 
import pandas as pd
import numpy as np
```

```python
et = pd.read_csv("data/earth_temp_updated.csv")
et = et.fillna("NULL")
et['Year'] = et['dt'].str.slice(0,4)
et['Year'] = et['Year'].astype(int)
et = et[(et['Year'] > 2004) & (et['Year'] < 2014)]
et
```

```python
sql_file = open('cpsc368_ddl4.sql', 'w')
```

```python
## USE NAMES FROM ER DIAGRAM NO CAPS UNDERSCORES BETWEEN

for index, row in et.iterrows():
    dt = str(row["dt"])
    avgtemp = str(row['AverageTemperature'])
    avgtu = str(row['AverageTemperatureUncertainty'])
    line = 'insert into temperature values (' + "'" + dt + "'" + ', ' + avgtemp + ', ' + avgtu + ');'
    sql_file.write(line +"\n")


```

```python
pc = pd.read_csv("data/BC_Power_Generation_updated.csv")
pc = pc[['source', 'year', 'value']]
pc = pc.rename(columns = {'year':'Year'})
pc = pc.fillna("NULL")
pc = pc[(pc['Year'] > 2004) & (pc['Year'] < 2014)]
pc
```

```python
sql_file.write("\n")
sql_file.write("--now add in power generated data")
```

```python
for index, row in pc.iterrows():
    source = str(row["source"])
    year = str(row['Year'])
    value = str(row['value'])
    line = 'insert into power_generated values (' + "'" + source + "'" ', ' + year + ', ' + value + ');'
    sql_file.write(line +"\n")
```

```python
pcons = pd.read_csv("data/BC_Power_Consumed_updated (1).csv")
pcons = pcons.rename(columns = {'year':'Year'})
pcons = pcons.fillna("NULL")
pcons = pcons[(pcons['Year'] > 2004) & (pcons['Year'] < 2014)]
pcons
```

```python
sql_file.write("\n")
sql_file.write("--now add in power consumed data")
```

```python
for index, row in pcons.iterrows():
    source = str(row["source"])
    year = str(row['Year'])
    value = str(row['value'])
    line = 'insert into power_consumed values (' + "'" + source + "'" + ', ' + year + ', ' + value + ');'
    sql_file.write(line +"\n")
```

```python
precip = pd.read_csv("data/Station_data_cleaned_updated.csv")
precip['MAX_TEMP'] = precip['MAX_TEMP'].replace(' None', np.nan)
precip['MAX_TEMP'] = precip['MAX_TEMP'].astype(float)
precip['ONE_DAY_PRECIPITATION'] = precip['ONE_DAY_PRECIPITATION'].replace(' None', np.nan)
precip['ONE_DAY_PRECIPITATION'] = precip['ONE_DAY_PRECIPITATION'].astype(float)
precip = precip.fillna("NULL")
precip = precip[(precip['Year'] > 2004) & (precip['Year'] < 2014)]
precip.dtypes
```

```python
sql_file.write("\n")
sql_file.write("--now add in precipitation data")
```

```python
for index, row in precip.iterrows():
    station = str(row["Station"])
    year = str(row['Year'])
    month = str(row['Month'])
    mintemp = str(row['MIN_TEMP'])
    maxtemp = str(row['MAX_TEMP'])
    pre = str(row['ONE_DAY_PRECIPITATION'])
    line = 'insert into precipitation values (' + "'" + station + "'" + ', ' + year + ', ' + month + ', ' + maxtemp + ', ' + mintemp + ', ' + pre +');'
    sql_file.write(line +"\n")
```

```python
pop_data = pd.read_csv("data/BC_Population_2005-2002_updated.csv")

pop_pg = pop_data.merge(pc, on = 'Year')
pop_pgc = pop_pg.merge(pcons, on = 'Year')
pop_pgc = pop_pgc.fillna("NULL")
pop_pgc = pop_pgc.drop_duplicates(['Year', 'Region'])
pop_pgc = pop_pgc[(pop_pgc['Year'] > 2004) & (pop_pgc['Year'] < 2014)]
pop_pgc
```

```python
sql_file.write("\n")
sql_file.write("--now add in population data")
```

```python
for index, row in pop_pgc.iterrows():
    sourcepg = str(row["source_x"])
    sourcepc = str(row['source_y'])
    year = str(row['Year'])
    region = str(row['Region'])
    munic = str(row['Municipality'])
    total = str(row['Total'])
    line = 'insert into population values (' + year + ', ' + region + ', ' + "'" + munic + "'" + ',' + "'" + total + "'" + ',' + "'" + sourcepg + "'" + ',' + year + ',' + "'" + sourcepc + "'" + ',' + year + ');'
    sql_file.write(line +"\n")
```
